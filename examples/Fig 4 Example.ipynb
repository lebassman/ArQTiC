```python
import sys
sys.path.append('/Users/cdp/Documents/GitHub/ArQTiC')
from arqtic.simulation_generator import Simulation_Generator
import numpy as np
```

```python
#create simulation object
sim_obj = Simulation_Generator("XY_input_randomized.txt")
```

```python
#connect to IBM
#if this is your first run enter your API key
#sim_obj.connect_IBM(api_key="your_key_here")
sim_obj.connect_IBM()
```

```python
sim_obj.generate_circuits()
```

```python
#run circuits on backend
sim_obj.run_circuits()
```

```python
noiseless_results=sim_obj.result_matrix
import matplotlib.pyplot as plt
plt.plot(noiseless_results)
#plt.ylim(-1,1)
#plt.xlim(0,sim_obj.steps)

plt.xlabel("Simulation Timestep")
plt.ylabel("Excitation Displacement")
plt.show()
```

```python
sim_obj.QCQS="QC"
sim_obj.device="ibmq_santiago"
sim_obj.run_circuits()
```

```python
qc_nonconstant=sim_obj.result_matrix

plt.plot(qc_nonconstant)
plt.plot(noiseless_results)
#plt.ylim(-1,1)
#plt.xlim(0,sim_obj.steps)

plt.xlabel("Simulation Timestep")
plt.ylabel("Excitation Displacement")
plt.show()
```

```python
sim_obj.constant_depth="True"
sim_obj.generate_circuits()
sim_obj.run_circuits()
```

```python
qc_constantdepth=sim_obj.result_matrix
plt.plot(qc_nonconstant,"r-.")
plt.plot(noiseless_results,"k")
plt.plot(qc_constantdepth,"b--")
#plt.ylim(-1,1)
#plt.xlim(0,sim_obj.steps)

plt.xlabel("Simulation Timestep",fontsize=14)
plt.ylabel("Excitation Displacement",fontsize=14)
plt.savefig("randomized_external_field.png")
plt.show()
```
